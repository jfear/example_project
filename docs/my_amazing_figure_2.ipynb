```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import pandas as pd
```

```python
# Load example data from the sklearn package
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target
df.target = df.target.map(dict(zip(range(4), iris.target_names)))
df
```

```python
# Wrangle data for easier plotting
dat = df.melt(id_vars="target", var_name="measurement")

# Make a panel plot
g = sns.FacetGrid(dat, col="measurement", col_wrap=2, hue="target")
g.map(sns.barplot, "target", "value", order=iris.target_names)
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Figure 2");
```
